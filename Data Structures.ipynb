## Lists (list)

Lists are ordered, mutable collections of elements that can contain any type of data, including other lists. Lists are extremely flexible and can be used to perform a wide variety of operations.


```python
list_only_numbers = [15, 2, 4.3, 5.25]
list_only_strings = ["banana", "apple", "pizza", "hamburger"]
list_mixed = ["cheese", 12, "mango", 1.533]
```

```python
list_only_numbers
```

```python
list_only_strings
```

```python
list_mixed
```

```python
# Sorting
sorted(list_only_numbers, reverse=False)
```

```python
sorted(list_only_numbers, reverse=True)
```

```python
sorted(list_only_strings, reverse=False)
```

```python
list_mixed
```

```python
# Can not sort mixed data type list!
sorted(list_mixed, reverse=False)
```

```python
# Index of a list
# [15, 2, 4.3, 5.25]
# starts from 0

print(list_only_strings[0])
print(list_only_strings[1])
print(list_only_strings[2])
print(list_only_strings[3])
print(f"Len of list (number of elements): {len(list_only_strings)}")
print(list_only_strings[4])

print("Only 4 element in list")
```

```python
# Change an element of a list
list_mixed[0] = "pepper"
print(list_mixed)
```

```python
# Appending an element to the list
list_only_numbers.append(9.83)
print(list_only_numbers)
```

```python
list_only_numbers.insert(2, 9.15) #insert to 3rd place an element
print(list_only_numbers)
```

```python
# Remove element from list
list_only_numbers.remove(9.15)
```

```python
list_only_numbers
```

```python
# Extend a list
list_only_numbers.extend([1, 2, 3])
```

```python
list_only_numbers
```

```python
# Slicing a list
print(list_mixed)
print(list_mixed[1:3]) #first, second
print(list_mixed[0:2])
print(list_mixed[-2:])
```

```python
# Concatenate two list
list_only_numbers + list_only_strings
```

```python
#  Nested list
nested_list = [[2, 5, 1], ["taco", "pizza", "wrap"]]
```

```python
print(nested_list[0])
print(nested_list[0][1])
print(nested_list[1])
print(nested_list[1][-1])
```

## Tuple

A tuple is a data structure in Python that consists of an ordered collection of elements which are immutable. Tuples offer several advantages, such as being faster and more secure than lists, as their elements cannot be modified.

```python
tuple_1 = (2.53, 6.554, 213.23, 32.4213)
tuple_2 = "apple", "microsoft", "ibm", "samsung", "apple", "nvidia"
tuple_3 = (4, 5, 6, 1)
```

```python
print(tuple_1)
print(tuple_2)
print(tuple_3)
```

```python
type(tuple_1)
```

```python
# Indexing

print(tuple_1[0])
print(tuple_1[1])
```

```python
# Slicing
print(tuple_2[2:4])
```

```python
# Concatenate tuples
tuple_1 + tuple_3
```

```python
# Len of a tuple
len(tuple_1)
```

```python
# Occurence of an element in a tuple
tuple_2.count("apple")
```

```python
tuple_3
```

```python
# Extract tuple elements to variables
number_1, number_2, number_3, number_4 = tuple_3
```

```python
print(number_1)
print(number_2)
print(number_3)
print(number_4)
```

```python
#  Convert to list
list(tuple_2)
```

## Dictionaries (dictionary)

Dictionaries contain key-value pairs, where each key is unique and associated with a value. Dictionaries are highly flexible and efficient, especially when quick lookups, insertions, or deletions are needed.


```python
dictionary_1 = {
    "Company": "Microsoft",
    "CEO": "Satya Nadella (since February 4, 2014)",
    "Founders": ["Bill Gates", "Paul Allen"],
    "Founded": "April 4, 1975, Albuquerque, New Mexico, USA",
    "President": "Satya Nadella",
    "Number of Employees": "221,000 (2023)",
    "Net Income": "73.4 billion USD (2023)",
    "Revenue": "175 billion USD (2021)"
}


dictionary_2 = {0: "Ready",
            1: "set",
            2: "go",
            3: "!"}
```

```python
# Change value at a specified key
dictionary_1["Revenue"] = "198.3 billion USD (2022)"
```

```python
dictionary_1
```

```python
# Reach values at keys
print(dictionary_1["Company"])
print(dictionary_1["Founded"])
print(dictionary_1["Net Income"])
```

```python
# Print out all the keys
dictionary_1.keys()
```

```python
# Print out all the values
dictionary_1.values()
```

```python
# All key-value pair (in tuples)
dictionary_1.items()
```

```python
# Change value
print(dictionary_2[2])
dictionary_2[2] = "wait"
print(dictionary_2[2])
```

```python
# New key-value pair
dictionary_1["Stock Price"] = 408
print(dictionary_1)
```

```python
# Remove key-value pair
del dictionary_1["President"]
print(dictionary_1)
```

```python
# Second method to remove a key-value pair
dictionary_1.pop("Net Income")
print(dictionary_1)
```

```python
# Concatenate two dictionary
print(dictionary_1)
dictionary_1.update(dictionary_2)
print(dictionary_1)
```

```python
dictionary_1[0]
```

```python
# Dict from tuple
tuple_list = [(0, "Apple"), (1, "Mango"), ("Melon", 2)]
```

```python
dict(tuple_list)
```

## Set

Sets are one of Python's built-in data structures that contain a collection of unique elements. Sets do not allow duplicate elements, and the order of elements is not defined. Sets are extremely useful for storing unique elements and quickly checking if an element is present in the set.


```python
set_1 = {5, 5, 2, 3, 4, 0, "cat"}
set_2 = {2, "cat", "dog", "hamster", "cat", "elephant"}
```

```python
set_1
```

```python
set_2
```

```python
# Adding new element to set
set_1.add(5)
print(set_1)
set_1.add(15)
print(set_1)
```

```python
set_1.add("cat")
```

```python
set_1
```

```python
#  Remove element
set_1.remove(0)
print(set_1)
```

```python
# Union of sets
set_1 | set_2 | set_2
```

```python
# Intersection of sets
set_1 & set_2
```

```python
# Subtraction of two set
set_1 - set_2
```

```python
 # Symmetrical subtraction of two set
 set_1 ^ set_2
```
