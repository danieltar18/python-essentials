### Pandas

Pandas is an open-source Python library primarily used for data manipulation and analysis. It provides two main data structures: Series and DataFrame.

- **Series**: A one-dimensional array-like object that can hold any data type.
- **DataFrame**: A two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns).

For more details, visit the official documentation: [pandas](https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html)

#### Dataframe manipulation

```python
# Slicing (columns)
dataframe.iloc[:, 0:3]
```

```python
# Slicing rows and cols
dataframe.iloc[-2:, 0:3]
```

```python
dataframe.tail()
```

```python
# Adding a new line (to the end)
new_line = {"Date": (pd.to_datetime(max(dataframe.Date)) +  pd.Timedelta(days=1)).date(), "Open": dataframe.Open.iloc[-7:].mean(),
            "High": dataframe.High.iloc[-7:].mean(), "Low": dataframe.Low.iloc[-7:].mean(),
            "Close": dataframe.Close.iloc[-7:].mean(), "Adj Close": dataframe["Adj Close"].iloc[-7:].mean(),
            "Volume": dataframe.Volume.iloc[-7:].mean()}
dataframe.loc[len(dataframe)] = new_line
```

```python
dataframe.tail()
```

```python
# New row (to the beginning)
new_line = {"Date": (pd.to_datetime(dataframe.Date.iloc[0]) -  pd.Timedelta(days=1)).date(), "Open": dataframe.Open.iloc[:7].mean(),
            "High": dataframe.High.iloc[:7].mean(), "Low": dataframe.Low.iloc[:7].mean(),
            "Close": dataframe.Close.iloc[:7].mean(), "Adj Close": dataframe["Adj Close"].iloc[:7].mean(),
            "Volume": dataframe.Volume.iloc[:7].mean()}
dataframe = pd.concat([pd.DataFrame([new_line]), dataframe], ignore_index=True)
dataframe.head()
```

```python
# New Col (on the end)
dataframe["etf_name"] = "VWCE"
dataframe.tail()
```

```python
dataframe.Date = pd.to_datetime(dataframe.Date)
```

```python
# New Col (insert to first place)
dataframe.insert(0, 'month', [date.strftime("%Y-%m") for date in dataframe.Date])
dataframe.head()
```

```python
# Delete rows
dataframe.drop([0, 1, 2, 3]) #inplace=True
```

```python
# Delete cols
dataframe.drop(["Open", "Volume"], axis=1)
```

```python
# Change a value in a specified row and col
dataframe.loc[3, "etf_name"] = "VWCE Vanguard"
```

```python
# Filter on string
dataframe[dataframe.etf_name == "vend"]
# or
dataframe[dataframe.etf_name.str.contains("Vanguard")]
```

```python
# Filter based on condition (number)
dataframe[dataframe.Open >= 98.25]
```

```python
# Filter on dates
dataframe[dataframe.Date >= "2024-01-18"]
```

```python
# Using AND & at condition filtering

dataframe[(dataframe.Date >= "2024-01-18") & (dataframe.Low < 110.5) & (dataframe.Volume > 75000)]
```

```python
# Using OR | at condition filtering

dataframe[(dataframe.Date >= "2024-05-18") | (dataframe.Open > 115.1) | (dataframe.Volume > 10e4)]
```

```python
# Filter on Nan/Nulls
dataframe[dataframe.isna().any(axis=1)] #axis=1 row, axis=0 column range
```

```python
# Replace Nans/Nulls
dataframe.fillna(0, inplace=True)
```

```python
# Drop NaN rows
dataframe.dropna(inplace=True)
```

```python
# Sort Values based on Volume
dataframe.sort_values("Volume", ascending=False, inplace=False)
```

### Aggregation

```python
dataframe.head()
```

```python
dataframe.loc[3, "etf_name"] = "VWCE"
```

```python
dataframe.groupby("etf_name").agg({"Volume": "sum", "High": "max", "Open": "min"})
```

```python
# Groupby
dataframe.groupby(["month", "etf_name"]).agg({"Volume": "sum", "Open": "min", "High": "max"})
```

```python
# Saving a dataframe
dataframe.to_csv("data/final.csv", index=False)
dataframe.to_excel("data/final.xlsx", index=False)
```

```python
import matplotlib.pyplot as plt
```

```python
dataframe.head()
```

```python
dataframe.set_index("Date", inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax.set_title("VWCE ETF Close Price last 3 Year", fontsize=12, fontweight="bold", fontstyle="italic")
ax.plot(dataframe.Close, color="#1f77b4", label="Close Price (€)")


ax.scatter(dataframe.index[dataframe.Close.argmax()], dataframe.Close.max(), marker="o", color="#DF1067", zorder=3)
ax.text(dataframe.index[dataframe.Close.argmax()], dataframe.Close.max(), f"Max value: {round(dataframe.Close.max())}", fontstyle="italic")


ax.scatter(dataframe.index[dataframe.Close.argmin()], dataframe.Close.min(), marker="o", color="r", zorder=3)
ax.text(dataframe.index[dataframe.Close.argmin()], dataframe.Close.min(), f"Min value: {round(dataframe.Close.min())}", fontstyle="italic")


ax.text(dataframe.index[0], dataframe.Close.mean()+3, f"Mean: {round(dataframe.Close.mean(), 2)}", fontweight="bold")
plt.axhline(dataframe.Close.mean(), label="Mean")


ax.set_xlabel("Date")
ax.set_ylabel("Close Price (€)")
ax.legend()
plt.xticks(rotation=90)

plt.show()
```

```python
period_start = "2020-01-01"
period_end = "2020-12-31"
fig, ax = plt.subplots(figsize=(10, 5))

dataframe_2020 = dataframe[period_start:period_end]

ax.set_title("VWCE ETF - Close Price vs. Volume 2020", fontsize=12, fontweight="bold", fontstyle="italic")
line1 = ax.plot(dataframe_2020.Close, color="#2ca02c", label="Close Price (€)")

plt.xticks(rotation=90)

ax2 = ax.twinx()
line2 = ax2.plot(dataframe_2020.Volume, color="#d62728", label="Volume")

ax.set_xlabel("Date")
ax.set_ylabel("Close Price (€)")
ax2.set_ylabel("Volume")

lns = line1 + line2
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, loc=2)

plt.show()
```
